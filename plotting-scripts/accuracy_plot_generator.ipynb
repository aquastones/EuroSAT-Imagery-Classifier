#### Imports

```python
import csv
import matplotlib.pyplot as plt
import seaborn as sb
```

#### Define Function

```python
def plot_csvs(filenames, labels, colors):
    plt.figure(figsize=(16, 9), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    for spine in ax.spines.values():
        spine.set_visible(False)

    for filename, label, color in zip(filenames, labels, colors):
        with open(filename) as f:
            data = [(int(r[0]), float(r[1])) for r in list(csv.reader(f))[1:]]

        sb.lineplot(x=[r[0] for r in data], y=[r[1] for r in data], label=label, color=color)

    plt.xlabel("Epochs", color="black")
    plt.ylabel("Values", color="black")
    plt.xticks(color="black")
    plt.yticks(color="black")
    plt.legend()
    plt.show()
```

#### Generation

```python
plot_csvs(
    ["csv/test_loss.csv", "csv/train_loss.csv", "csv/accuracy.csv"],
    ["Test Loss", "Train Loss", "Accuracy"],
    ["red", "blue", "green"]
)
```
